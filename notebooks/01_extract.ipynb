```python
import sys
from pathlib import Path
import json
import re

ROOT = Path.cwd().parents[0]
sys.path.append(str(ROOT / "src"))

from api.pokeapi import fetch_endpoint, save_json

pokemon_list = fetch_endpoint("pokemon?limit=2000")["results"]

id_pattern = re.compile(r"/pokemon/(\d+)/?$")

for entry in pokemon_list:
    match = id_pattern.search(entry["url"])
    if not match:
        continue
    
    poke_id = int(match.group(1))

    # Skip form IDs (species only go to 1025)
    if poke_id > 1025:
        continue

    mon = fetch_endpoint(f"pokemon/{poke_id}")
    save_json(mon, ROOT / f"data_raw/pokemon/{poke_id}.json")

    species = fetch_endpoint(f"pokemon-species/{poke_id}")
    save_json(species, ROOT / f"data_raw/species/{poke_id}.json")
```
